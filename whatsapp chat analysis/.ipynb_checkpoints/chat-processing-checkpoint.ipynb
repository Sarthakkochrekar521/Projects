```python
import re # re = regular expression
import pandas as pd
```

```python
f = open('WhatsApp Chat with TE COMP C Official 23-24.txt','r',encoding='utf-8')
# opening textfile using fliehandling just make sure the encoding is utf-8
# r = read mode
```

```python
data = f.read() #read file content as string & stored in 'data'
```

```python
print(data) # print the 'data'
```

```python
print(type(data))
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s' 
#
# to segregate the data & time and user & his message into columns, we made a
# regular expression string using tools like 'regex' on the basis of message
# pattern which looks like :
# d{1,2}- digit which contains one or two numbers, eg: 8 or 08 (donotes day)
# d{1,2}- digit which contains one or two numbers, eg: 8 or 08 (donotes month)
# d{2,4}- digit which contains two or four numbers, eg: 22 or 2022 (donotes year)
# \s\ - space
# d{1,2}- digit which contains one or two numbers, eg: 8 or 18 (denotes hour)
# : - colon
# d{2}- digit which contains only two numbers, eg: 08 or 18 (denotes minute)
# - - hyphen
```

```python
messages = re.split(pattern,data)[1:] # listing the messages by splitting pattern from data
messages
```

```python
dates = re.findall(pattern,data) # extracting the dates and time
dates
```

```python
df = pd.DataFrame({'user_ message':messages, 'message_date':dates}) # formating the data to dataframes

# convert messages_date type
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %H:%M - ')

df.rename(columns={'message_date':'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
# separate user and messages
users = [] # making a userlist
messages =[] # making a messagelist
for message in df['user_ message']: #for loop on user_message column
    entry = re.split('([\w\W]+?):\s',message) # splitting on pattern i.e
    # if there is a colon between name and string, then
    if entry[1:]: # user name
        users.append(entry[1]) # insert all names in userlist
        messages.append(entry[2]) # insert all string in messagelist
    #if not, then
    else:
        users.append('group_notification') # consider it as notifcation
        messages.append(entry[0]) # otherwise insert as it is in messages
        
# inserting new columns
df['user'] = users
df['message'] = messages
df.drop(columns=['user_ message'], inplace=True)

df.head()
```

```python
df['year'] = df['date'].dt.year # extracting year from date
```

```python
period = [] # making a periodlist
for hour in df[['day_name','hour']]['hour']:#to make a pattern 'hour-hour'
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
# Sentimental Analysis
#1.remove group notification
senti_mess = df[df['user'] != 'group_notification']
#2.remove 'Media ommitted' messages
senti_mess = senti_mess[senti_mess['message'] != '<Media omitted>\n']
#3.remove empty messages
senti_mess = senti_mess[senti_mess['message'].str.strip().astype(bool)]
#4.drop unecesary columns
senti_mess.drop(['year','month','day','hour','minute','month_num','only_date','day_name','period'],axis=1,inplace=True)
#5.reseting index
sen = senti_mess.reset_index(drop=True)

sen.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sen['vader_polarity'] = sen['message'].apply(lambda x: SentimentIntensityAnalyzer().polarity_scores(x)['compound'])
```

```python
def categorize_vader_sentiment(score):
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

```python
sen['vader_sentiment'] = sen['vader_polarity'].apply(categorize_vader_sentiment)
```

```python
sen.head(20)
```

```python
print(sen)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
sen_test = pd.read_csv("chat_dataset.csv")
sen_test.head()
```

```python
X = sen['message']
Y = sen['vader_sentiment']
#spilting the dataset into training and testing set
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#applying TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # You can adjust max_features
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# Train an SVM classifier
svm_classifier = SVC(kernel='linear')  # Here,we are using svm linear classifier
svm_classifier.fit(X_train_tfidf, Y_train)
```

```python
# making predictions on the test set
Y_pred = svm_classifier.predict(X_test_tfidf)
```

```python
sen['Sentiment'] = pd.DataFrame(Y_pred)
```

```python
sen.head()
```

```python
sen.groupby('Sentiment').describe()
```

```python
# Evaluate the model
accuracy = accuracy_score(Y_test, Y_pred)
classification_rep = classification_report(Y_test, Y_pred)

print("Accuracy:", accuracy)
print("\nClassification Report:\n", classification_rep)
```
